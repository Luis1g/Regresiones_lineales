## **Regresión lineal por mínimos cuadrados**

### Importamos las librerías.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### **1. Generamos los datos de ejemplo para nuestra simulación**

Suponemos que X son "Horas de estudio" y Y es "Calificación"

- `np.random.seed(n)` nos ayuda a generar datos pseudoaleatorios con una semilla `n` específica. En nuestro caso usamos 42 solo por convención.
- `np.random.rand(l,m)` crea una matriz con l filas y m columnas de números aleatorios en el rango de [0, 1).

```python
np.random.seed(42)
X = 2 * np.random.rand(100,1)
Y = 4 + 3 * X + np.random.rand(100, 1)
```

Con ayuda de `flatten()` aplanamos X e Y. Convertimos un array multidimensional en un array unidimensional.

```python
x_val = X.flatten()
y_val = Y.flatten()
```

### **2. Preparación de los datos para la regresión lineal**

Calculamos el promedio de X y Y

```python
x_prome = np.mean(x_val)
y_prome = np.mean(y_val)
```

Calculamos $\beta_1$ y $\beta_0$

$$
\begin{equation}
\beta_1  = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i-\bar{y})}{\sum_{i=1}^n(x_i -\bar{x})^2}
\end{equation}
$$

$$
\begin{equation}
    \beta_0 = \bar{y} - \beta_1 \bar{x}
\end{equation}
$$

```python
numerador = 0
denominador = 0

for i in range(len(x_val)):
    numerador += (x_val[i] - x_prome) * (y_val[i] - y_prome)
    denominador += (x_val[i] - x_prome) ** 2

beta_1 = numerador / denominador
beta_0 = y_prome - (beta_1 * x_prome)
```

### **3. Resultados y predicción**

```python
print(f"La Intersección (beta_0) calculada es: {beta_0:.4f}")
print(f"La Pendiente (beta_1) calculada es: {beta_1:.4f}")
print(f"\nLa ecuación de la recta es: y = {beta_0:.2f} + {beta_1:.2f}x")

# Hacemos la predicción para dibujar la línea
y_pred = beta_0 + beta_1 * x_val
```

### **4. Visualización**

```python
plt.scatter(x_val, y_val, color='blue', alpha=0.5, label='Datos Reales')
plt.plot(x_val, y_pred, color='red', linewidth=2, label='Línea de Regresión (OLS)')
plt.xlabel('Variable X (independiente)')
plt.ylabel('Variable Y (dependiente)')
plt.legend()
plt.title('Regresión Lineal OLS')
plt.show()
```
